```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_optimal_skills = pd.read_csv('../csv_files_datapull/5_optimal_skills.csv')
df_optimal_skills.head()
```

### Optimal Skill Salary Table

```python
# Name Variables
optimal_skill = df_optimal_skills['skills']
skill_average_salary = df_optimal_skills['avg_salary_$']
bar_color = '#8E4585'

plt.figure(figsize=(10, 6))
plt.bar(optimal_skill, skill_average_salary, color=bar_color)
plt.title('Most Optimal Skills to Learn based on Salary (2025)')
plt.xlabel('Skill')
plt.ylabel('Average Salary ($)')
plt.ylim(0, 135000)
plt.yticks([0, 25000, 50000, 75000, 100000, 125000, 150000])  # Add specific ticks
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Optimal Skill Count Table

```python
# Name Variables
optimal_skill = df_optimal_skills['skills']
skill_demand_count = df_optimal_skills['demand_count']
bar_color = '#8E4585'

plt.figure(figsize=(10, 6))
plt.bar(optimal_skill, skill_demand_count, color=bar_color)
plt.title('Most Optimal Skills to Learn based on Demand (2025)')
plt.xlabel('Skill')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
