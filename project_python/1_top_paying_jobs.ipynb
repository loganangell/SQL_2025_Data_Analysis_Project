```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the CSV file into a DataFrame
df_top_paying_jobs = pd.read_csv('../csv_files_datapull/1_top_paying_jobs.csv')
df_top_paying_jobs.head()
```

```python
# Sort by salary descending
df_sorted = df_top_paying_jobs.sort_values(by='average_yearly_salary_$', ascending=False)

# Reverse the order for barh so highest is at the top
df_sorted = df_sorted[::-1]

# Name variables
salary_data = df_sorted['average_yearly_salary_$']
job_title = df_sorted['full_job_title']
bar_color = '#8E4585'

plt.figure(figsize=(10, 6))
plt.barh(job_title, salary_data, color=bar_color)
plt.xlabel('Average Yearly Salary ($)')
plt.title('Average Salary Distribution for Top 10 Paying Data Jobs in 2024')
plt.tight_layout()
plt.show()
```
