```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the CSV file into a DataFrame
df_top_paying_skills = pd.read_csv('../csv_files_datapull/4_top_paying_skills.csv')
df_top_paying_skills.head(10)
```

```python
# Sort the data from largest to smallest salary
df_sorted = df_top_paying_skills.sort_values('avg_salary_$', ascending=False)

# Name Variables
skill_name = df_sorted['skills']
average_salary = df_sorted['avg_salary_$']
bar_color = '#8E4585'

plt.figure(figsize=(10, 6))
plt.bar(skill_name, average_salary, color=bar_color)
plt.title('Top Paying Skills based on Salary (2025)')
plt.xlabel('Skill')
plt.ylabel('Average Salary ($)')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Prevent label cutoff
plt.show()
```
